# Covid spread relation with position

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('data/data_spread_rate_reproduction.csv')
```

```python
plt.figure(figsize=(10, 10))
plt.rcParams.update({'font.size': 15})
latitude = abs(data['Latitude'])
plt.scatter(latitude, data['Mortality Rate'], c='blue', alpha=0.5)
plt.xlabel('Absolute Latitude')
plt.ylabel('Mortality Rate')
plt.title('Mortality Rate vs Absolute Latitude')
plt.show()
```

```python
correlation = data['Mortality Rate'].corr(latitude)
print('Correlation between Mortality Rate and Absolute Latitude:', correlation)
```
